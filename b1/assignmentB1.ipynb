```python
import pandas as pd
```

```python
df=pd.read_csv("dataset_Facebook - dataset_Facebook.csv")
```

```python
df
```

```python
subset1=df[['Type', 'Total Interactions', 'Paid']]
subset1
```

```python
subset2=df[['Category', 'share']]
subset2
```

```python
subset3=df[['Post Month', 'comment']]
subset3
```

```python
subset4=df[['Lifetime Post Consumers', 'comment']]
subset4
```

#merge data

```python
merge1=pd.merge(subset2, subset1, left_on='share', right_on='Total Interactions')
merge1
```

```python
merge2=pd.merge(subset4, subset3, on='comment' )
merge2
```

#sort data

```python
merge1.sort_values(by=['share'],ascending=False)
```

#Transposing Matrix

```python
merge1.transpose()
```
